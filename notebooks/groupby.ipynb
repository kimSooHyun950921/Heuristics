```python
grp = df.groupby([0])
grp
```

```python
import time
```

```python
count = 0
all_list = []
stime = time.time()
for name, group in grp:
    all_list.append((str(name), count))
    for k in group[1].to_list():
        all_list.append((str(k), count))
        if len(all_list) % 10000 == 0:
            print(f"length: {len(all_list)}, value:{all_list[-1]}")
            etime = time.time()
            print(f"current time: {etime - stime}")
    if count % 10000 == 0:
        etime = time.time()
        print(f"GROUP: {count} END")
        print(f"current time: {etime - stime}")
    count += 1   
```
