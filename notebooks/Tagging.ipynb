```python
mi_group = mi_df.groupby('number')
mi_dict = []
#{'Group':1234, 'Tag':[], 'intxNum':12, 'outTx':13, 'InValue':13, 'OutValue':123}
```

```python
cluster_list = []
count = 0
with open('/home/dnlab/DataHDD/all_info.csv', 'w', newline='') as csvfile:
    fieldnames = ['group', 'Category','num_in_tx','num_out_tx', \
                  'in_value', 'out_value', 'balance', \
                  'smallest_addr', 'ClusterName', 'len']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for cluster_number, addr_group in mi_group:
        if len(mi_group) > 100000:
            print("OVER 100000 the length of addr_list:", len(addr_group))
        addr_dict = {'group':cluster_number, 'Category':[], 'num_in_tx':0, \
                     'num_out_tx':0, 'in_value':0, 'out_value':0, 'balance':0,\
                     'smallest_addr':addr_group.reset_index().iloc[0]['addr'],\
                     'ClusterName':[],'len': len(addr_group)}
        addr_list = addr_group['addr'].to_list()


        if len(addr_list) > 10000:
            print("query start", len(addr_list))
            
        addr_dict['num_in_tx'] = dq.count_all_tx_in(addr_list)[0]
        addr_dict['num_out_tx'] = dq.count_all_tx_out(addr_list)[0]
        addr_dict['in_value'] = dq.sum_all_invalue(addr_list)[0]
        addr_dict['out_value'] = dq.sum_all_outvalue(addr_list)[0]
        addr_dict['balance'] = addr_dict['out_value'] - addr_dict['in_value']
        
        if len(addr_list) > 10000:
            print("join_start:", len(addr_list))   
        join_df = pd.merge(addr_group, root_addr_df,how = 'inner', on=['addr'])

        category = join_df['Category'].to_list()
        addr_dict['Category'].append(category) 

        cluster_name = join_df['ClusterName'].to_list()
        addr_dict['ClusterName'].append(cluster_name) 

        writer.writerow(addr_dict)
        if count % 100000== 0:
            print(count)
        count += 1
```
